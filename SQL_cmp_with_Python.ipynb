# **JOIN**

```python
df1 = pd.DataFrame({"key": ["Mon", "Tue", "Wed", "Sun"], "value": np.random.randint(4)})
df2 = pd.DataFrame({"key": ["Mon", "Mon", "Wed", "S"], "value": np.random.randint(4)})

```

## INNER JOIN

```python
# INNER JOIN
'''
SELECT *
FROM df1
INNER JOIN df2
  ON df1.key = df2.key;
  '''
pd.merge(df1, df2, on="key")
```

##LEFT OUTER JOIN

```python
'''
SELECT *
FROM df1
LEFT OUTER JOIN df2
  ON df1.key = df2.key;
  '''
pd.merge(df1, df2, on="key", how="left")
```

##RIGHT JOIN

```python
'''
SELECT *
FROM df1
RIGHT OUTER JOIN df2
  ON df1.key = df2.key;
  '''
pd.merge(df1, df2, on="key", how="right")
```

##FULL JOIN

```python
'''
SELECT *
FROM df1
FULL OUTER JOIN df2
  ON df1.key = df2.key;
  '''
pd.merge(df1, df2, on="key", how="outer")
```

##UNION

```python
df1 = pd.DataFrame(
    {"city": ["Delhi", "San Francisco", "New York City"], "rank": range(1, 4)})
df2 = pd.DataFrame(
    {"city": ["Delhi", "Boston", "Los Angeles"], "rank": [1, 4, 5]})
```

```python
'''
SELECT city, rank
FROM df1
UNION ALL
SELECT city, rank
FROM df2;
'''
pd.concat([df1, df2])
```

```python
'''
SELECT city, rank
FROM df1
UNION
SELECT city, rank
FROM df2;
'''
pd.concat([df1, df2]).drop_duplicates()
```

##Top n rows per group
##### Oracle's ROW_NUMBER() analytic function

```python
'''
SELECT * FROM (
  SELECT
    t.*,
    ROW_NUMBER() OVER(PARTITION BY day ORDER BY total_bill DESC) AS rn
  FROM tips t
)
WHERE rn < 3
ORDER BY day, rn;
'''
(
    tips.assign(
        rn=tips.sort_values(["total_bill"], ascending=False)
        .groupby(["day"])
        .cumcount()
        + 1
    )
    .query("rn < 3")
    .sort_values(["day", "rn"])
)
```

#####the same using rank(method='first') function

```python
(
    tips.assign(
        rnk=tips.groupby(["day"])["total_bill"].rank(
            method="first", ascending=False
        )
    )
    .query("rnk < 3")
    .sort_values(["day", "rnk"])
)
```

#####Oracle's RANK() analytic function

```python
'''
SELECT * FROM (
  SELECT
    t.*,
    RANK() OVER(PARTITION BY sex ORDER BY tip) AS rnk
  FROM tips t
  WHERE tip < 2
)
WHERE rnk < 3
ORDER BY sex, rnk;
'''
(
    tips[tips["tip"] < 2]
    .assign(rnk_min=tips.groupby(["sex"])["tip"].rank(method="min"))
    .query("rnk_min < 3")
    .sort_values(["sex", "rnk_min"])
)
```
